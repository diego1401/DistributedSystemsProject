```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
title = "log_test_seq_vs_graph_part20000.txt"
filepath = os.path.abspath(title)
lines = open(filepath).readlines()
num_threads = []
time1 = []
time_CS = []
time_Com = []
for line in lines[2:]:
    line = line.strip().split(",")
    num_threads.append(line[0])
    time1.append(float(line[1]))
    time_CS.append(float(line[2]))
    time_Com.append(float(line[3]))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(num_threads,time1)
# plt.plot(num_threads,Seq,label="Sequential")
plt.xlabel("Number of threads")
plt.ylabel("Time in seconds")
title = "Time of Execution of the Parallel SSSP based on Graph-Partitioning and Iterative Correcting and the Sequetial algorithm"
plt.title(title)
# plt.legend()
plt.savefig(title)
plt.show()
```

```python
title = "log_test_seq_vs_graph_part20000bis.txt"
filepath = os.path.abspath(title)
lines = open(filepath).readlines()
num_threads = []
time = []
time_CS = []
time_Com = []
Seq = []
for line in lines[2:]:
    line = line.strip().split(",")
    num_threads.append(line[0])
    time.append(float(line[1]))
    time_CS.append(float(line[2]))
    time_Com.append(float(line[3]))
    Seq.append(float(line[4]))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(num_threads,time,label="Parallel")
plt.plot(num_threads,Seq,label="Sequential")
plt.xlabel("Number of threads")
plt.ylabel("Time in seconds")
title = "Time of Execution of the Parallel SSSP based on Graph-Partitioning and Iterative Correcting and the Sequetial algorithm"
plt.title(title)
plt.legend()
plt.savefig(title)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(num_threads,time1,label="10%")
plt.plot(num_threads,time,label="1%")
plt.xlabel("Number of threads")
plt.ylabel("Time in seconds")
title = "Time of Execution of the Parallel SSSP based on Graph-Partitioning and Iterative Correcting with different Sparsity"
plt.title(title)
plt.legend()
plt.savefig(title)
plt.show()
```

```python
ratio_CS = [time_CS[i]/(time_CS[i]+time_Com[i]) for i in range(len(num_threads))]
ratio_Com = [time_Com[i]/(time_CS[i]+time_Com[i]) for i in range(len(num_threads))]
```

```python
stop = 7
plt.figure(figsize=(10, 6))
plt.plot(num_threads[:stop],ratio_CS[:stop],label="Cost of Computation")
plt.plot(num_threads[:stop],ratio_Com[:stop],label="Cost of Communication")
plt.xlabel("Number of threads")
plt.ylabel("Percentage of impact on final time")
title = "Comparison of the Computation and Communication step of the algorithm"
plt.title(title)
plt.legend()
plt.savefig(title)
plt.show()
```
