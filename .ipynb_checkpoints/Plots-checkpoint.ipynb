```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
title = "log_test_seq_vs_graph_part_20000.txt"
filepath = os.path.abspath(title)
lines = open(filepath).readlines()
num_threads = []
time = []
time_CS = []
time_Com = []
for line in lines[2:]:
    line = line.strip().split(",")
    num_threads.append(line[0])
    time.append(line[1])
    time_CS.append(line[2])
    time_Com.append(line[3])
    
```

```python
plt.figure(figsize=(12, 6))
plt.plot(num_threads,time)
plt.xlabel("Number of threads")
plt.ylabel("Time")
title = "Time of Execution of the Parallel SSSP based on Graph-Partitioning and Iterative Correcting"
plt.title(title)
# plt.legend()
plt.savefig(title)
plt.show()
```
